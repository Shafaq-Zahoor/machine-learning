```python
def plot_clusters(data_scaled):
    kmeans = KMeans(n_clusters=4, init='k-means++')
    kmeans.fit(data_scaled)
    labels = kmeans.predict(data_scaled)
    centroids = kmeans.cluster_centers_
    start_time = time.time() 
    end_time = time.time()
    # code to plot clusters with labels and centroids
    plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=labels, s=50, cmap='viridis')
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.5)
    
    # code to add title and text
    plt.title('Clusters found by KMeans', fontsize=24)
    plt.text(-0.5, 0.7, 'Clustering took {:.2f} s'.format(end_time - start_time), fontsize=14)
    plt.axis('off')
    plt.show()

# call the function with scaled_data
plot_clusters(data_scaled)
```

```python
from sklearn.cluster import AffinityPropagation

from sklearn.metrics import pairwise_distances

def plot_clusters(data_scaled):
  median_distance = np.median(pairwise_distances(data_scaled))
  preference = -median_distance ** 2
  damping = 0.9
  max_iter = 500
  clusterer = AffinityPropagation(preference=preference, damping=damping, max_iter=max_iter)
  clusterer.fit(data_scaled)
  labels = clusterer.predict(data_scaled)
  
  start_time = time.time() 
  end_time = time.time()
    # code to plot clusters with labels and centroids
  plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=labels, s=50, cmap='viridis')
  
    
    # code to add title and text
  plt.title('Clusters found by Affinity Propagation', fontsize=24)
  plt.text(-0.5, 0.7, 'Clustering took {:.2f} s'.format(end_time - start_time), fontsize=14)
  plt.axis('off')
  plt.show()

# call the function with scaled_data
plot_clusters(data_scaled)
```

```python
from sklearn.cluster import MeanShift, estimate_bandwidth



def plot_clusters(data_scaled):
  bandwidth = estimate_bandwidth(data_scaled, quantile=0.1)
  ms = MeanShift(bandwidth=bandwidth)
  ms.fit(data_scaled)
  labels = ms.predict(data_scaled)
  
  start_time = time.time() 
  end_time = time.time()
    # code to plot clusters with labels and centroids
  plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=labels, s=50, cmap='viridis')
  
    
    # code to add title and text
  plt.title('Clusters found by Mean Shift', fontsize=24)
  plt.text(-0.5, 0.7, 'Clustering took {:.2f} s'.format(end_time - start_time), fontsize=14)
  plt.axis('off')
  plt.show()

# call the function with scaled_data
plot_clusters(data_scaled)
```

```python
from sklearn.cluster import SpectralClustering



def plot_clusters(data_scaled):
  n_clusters = 4
  affinity = 'laplacian'
  gamma = 1.0
  sc = SpectralClustering(n_clusters=n_clusters, affinity=affinity, gamma=gamma)
  
  sc.fit(data_scaled)
  labels = sc.fit_predict(data_scaled)
  
  start_time = time.time() 
  end_time = time.time()
    # code to plot clusters with labels and centroids
  plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=labels, s=50, cmap='viridis')
  
    
    # code to add title and text
  plt.title('Clusters found by Spectral Clustering', fontsize=24)
  plt.text(-0.5, 0.7, 'Clustering took {:.2f} s'.format(end_time - start_time), fontsize=14)
  plt.axis('off')
  plt.show()

# call the function with scaled_data
plot_clusters(data_scaled)
```

```python
from sklearn.cluster import AgglomerativeClustering



def plot_clusters(data_scaled):
  n_clusters = 5
  linkage = 'ward'  # the linkage criterion to use (ward, complete, or average)
  affinity = 'euclidean'  # the distance metric to use
  clustering = AgglomerativeClustering(n_clusters=n_clusters, linkage=linkage, affinity=affinity)
  
  clustering.fit(data_scaled)
  labels = clustering.fit_predict(data_scaled)
  
  start_time = time.time() 
  end_time = time.time()
    # code to plot clusters with labels and centroids
  plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=labels, s=50, cmap='viridis')
  
    
    # code to add title and text
  plt.title('Clusters found by Agglomerative Clustering', fontsize=24)
  plt.text(-0.5, 0.7, 'Clustering took {:.2f} s'.format(end_time - start_time), fontsize=14)
  plt.axis('off')
  plt.show()

# call the function with scaled_data
plot_clusters(data_scaled)
```
